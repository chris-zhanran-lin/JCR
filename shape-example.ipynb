```python
import numpy as np
import sympy
import random
import math
import matplotlib.pyplot as plt
from sympy import *
import scipy
from scipy.stats import chi2  

np.random.seed(4)
n = 101
y = np.ones((n,1))
theta0 = 1
for i in range(n):
    y[i] = theta0 + np.random.normal(0)
theta_min = -2
theta_max = 4
grid_num = 1001
thetagrid = np.linspace(theta_min, theta_max, grid_num)
y_u1 = np.linspace(theta_min, theta_max, grid_num)
y_l1 = np.linspace(theta_min, theta_max, grid_num)

y_u2 = np.linspace(theta_min, theta_max, grid_num)
y_l2 = np.linspace(theta_min, theta_max, grid_num)

y_u3 = np.linspace(theta_min, theta_max, grid_num)
y_l3 = np.linspace(theta_min, theta_max, grid_num)

y_u4 = np.linspace(theta_min, theta_max, grid_num)
y_l4 = np.linspace(theta_min, theta_max, grid_num)

y_u5 = np.linspace(theta_min, theta_max, grid_num)
y_l5 = np.linspace(theta_min, theta_max, grid_num)

y1 = y[:(n-1),]  
n1 = n-1
w = 0.00001
for i in range(len(thetagrid)):
    y_u1[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .95,0,1)
    y_l1[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .05,0,1)

w = 1
for i in range(len(thetagrid)):
    y_u2[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .95,0,1)
    y_l2[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .05,0,1)

w = - n1
for i in range(len(thetagrid)):
    y_u3[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .95,0,1)
    y_l3[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .05,0,1)


w = 0.1 * n
for i in range(len(thetagrid)):
    y_u5[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .95,0,1)
    y_l5[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .05,0,1)    

w = 10000
for i in range(len(thetagrid)):
    y_u4[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .95,0,1)
    y_l4[i] = (1+(n1/w))*thetagrid[i] - (1/w)*sum(y1) + sqrt(1 + (n1/(w**2)))*scipy.stats.norm.ppf( .05,0,1)
```

```python
fig, ax = plt.subplots(dpi=300,figsize=(5,4.7))

ax.plot(thetagrid, y_u4,label='$ω→∞$',linewidth=1.5,c='darkorange',marker='1',markevery=50,markersize=5)
ax.plot(thetagrid, y_l4, c='darkorange',linewidth=1.5,marker='1',markevery=50,markersize=5)


ax.plot(thetagrid, y_u5,label='$ω= 0.1n$',linewidth=1.5,c='lightskyblue',marker='2',markevery=50,markersize=5)
ax.plot(thetagrid, y_l5, c='lightskyblue',linewidth=1.5,marker='2',markevery=50,markersize=5)

ax.plot(thetagrid, y_u2,label='$ω=1$',linewidth=1.5,c='plum',marker='v',markevery=50,markersize=5)
ax.plot(thetagrid, y_l2, c='plum',linewidth=1.5,marker='v',markevery=50,markersize=5)

ax.plot(thetagrid, y_u1,label='$ω→0$',linewidth=1.5,c='slateblue',marker='D',markevery=50,markersize=5)
ax.plot(thetagrid, y_l1, c='slateblue',linewidth=1.5,marker='D',markevery=50,markersize=5)

ax.plot(thetagrid, y_u3,label='$ω=-n$',linewidth=1.5,c='gold',marker='d',markevery=50,markersize=5)
ax.plot(thetagrid, y_l3, c='gold',linewidth=1.5,marker='d',markevery=50,markersize=5)

plt.ylabel('$y_{\mathrm{te}}$',size=15)
plt.xlabel('$θ$',size=15)
plt.xlim(-2,4)
plt.ylim(-24,26)
plt.legend(loc='upper left',fontsize=12)
plt.savefig('shape-example.png',bbox_inches = 'tight')
plt.show()
```
