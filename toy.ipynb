```python
import numpy as np
import sympy
import random
import math
import matplotlib.pyplot as plt
from sympy import *
import scipy

theta_min = -2
theta_max = 4
grid_num = 1001
thetagrid = np.linspace(theta_min, theta_max, grid_num)

jcr_u = np.linspace(theta_min, theta_max, grid_num)
jcr_l = np.linspace(theta_min, theta_max, grid_num)

xte = 1

for i in range(len(thetagrid)):
    jcr_u[i] = np.dot(xte,thetagrid[i]) + scipy.stats.norm.ppf( .95,0,1)
    jcr_l[i] = np.dot(xte,thetagrid[i]) + scipy.stats.norm.ppf( .05,0,1)
```

```python
fig, ax = plt.subplots(dpi=300,figsize=(5,4.7))

ax.fill_between(thetagrid, jcr_l, jcr_u, alpha=.5, linewidth=0,color='pink',label='JCR')

plt.ylabel('$y_0$',size=15)
plt.xlabel('$θ$',size=15)
plt.xlim(-2,4)
plt.ylim(-2,4)
plt.legend(loc='lower right',fontsize=11.7,framealpha=0.9)
plt.savefig('toy.png',bbox_inches = 'tight')
plt.show()
```
